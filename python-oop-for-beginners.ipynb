#  What is Programming?

Programming is the process of giving instructions to a computer so it can perform specific tasks.

Think of a computer as a very intelligent machine that follows instructions exactly as they are written.

These instructions are written using programming languages like:

- Python
- Java
- C++
- JavaScript

Without instructions, a computer cannot make decisions on its own.

In simple words,

> Programming is the process of telling a computer **what to do** and **how to do it**.

```python
print("Hello, World!")
```

The computer reads the instruction above and prints the text on the screen.

This is where **Object-Oriented Programming (OOP)** becomes useful.

#  What is Object-Oriented Programming (OOP)?



Object-Oriented Programming (OOP) is a programming style that organizes code using **objects**.

Instead of writing one long program, we divide it into smaller, reusable parts called **objects**.

**Each object contains:**

- Data (called attributes)
- Actions (called methods)



This makes programs easier to understand, maintain, and reuse.

## 🚗 Real-Life Example

Think about a car.

A car has:

**Attributes:**
- Color
- Brand
- Speed

**Actions**:
- Start
- Stop
- Accelerate
- Brake

**In OOP, we represent the car as an object.**

The object stores both the information (attributes) and the actions (methods).

This is exactly how OOP models real-world things.

#  What is a Class?

## What is a Class?


A **Class** is a blueprint or template used to create objects.

It defines what data (attributes) and actions (methods) an object will have.

Think of a class as a design, not the actual object.




---

## Why do we need a Class?

Without a class, we would write the same code repeatedly.

A class allows us to create multiple objects with the same structure.

---

## Real-Life Example

Imagine a house blueprint.

The blueprint is not a real house.

It only describes how houses should be built.

Similarly, a class is a blueprint for creating objects.

---

## Syntax

```python
class Student:
    pass
```

The keyword **`class`** is used to create a class.

**`Student`** is the name of the class.

**`pass`** means the class is currently empty.

#  What is an Object?


An **Object** is a real instance of a class.

If a class is a blueprint, then an object is the actual thing created from that blueprint.

---

## Why do we need Objects?

Objects allow us to use the class in real programs.

Each object can store different data.

---

## Real-Life Example

**Class** **→** Car Design

**Object** **→** Toyota Corolla

**Object** **→** Honda Civic




Both are created from the same blueprint but have different information.

```python
class Student:
    pass

student1 = Student()
student2 = Student()

print(student1)
print(student2)
```

### Creating Objects

```python
class Student:
    pass

s1 = Student()
s2 = Student()
s3 = Student()

print(s1)
print(s2)
print(s3)
```

#  Instance Variables

## What are Instance Variables?

Instance variables store information about an object.

Each object has its own copy of these variables.

---

## Why do we use them?

They allow different objects to store different values.

**Example:**

Student 1 → Ali

Student 2 → Sara

Each object stores its own data.

```python
class Student:

    def __init__(self, name, age):
        self.name = name
        self.age = age

s1 = Student("Ali", 20)
s2 = Student("Sara", 22)

print(s1.name)
print(s2.name)
```

#  Methods

## What is a Method?

A method is a function inside a class.

Methods define what an object can do.

---

## Why do we use Methods?

Instead of writing code repeatedly, we place actions inside methods.

Objects can then perform those actions whenever needed.

```python
class Student:

    def greet(self):
        print("Hello Student!")

s1 = Student()

s1.greet()
```

#  Self

## What is self?

`self` refers to the current object.

Python automatically sends the current object as the first argument whenever a method is called.

---

## Why is self needed?

Without `self`, Python would not know which object's data should be used.

Think of `self` as saying:

"This object's information."

```python
class Student:

    def greet(self):
        print("Hello!")

s1 = Student()

s1.greet()
```

#  Constructor (__init__)

## What is a Constructor?

A constructor is a special method that runs automatically when an object is created.

Its main purpose is to initialize the object's data.

---

## Why do we use it?

It saves time by setting values automatically during object creation.

You don't need to assign values separately later.

```python
class Student:

    def __init__(self, name):
        self.name = name

s1 = Student("Ali")

print(s1.name)
```

#  Mini Project

**Let's combine everything we learned.**

```python
class Student:

    def __init__(self, name, age, marks):
        self.name = name
        self.age = age
        self.marks = marks

    def display(self):
        print(f"Name  : {self.name}")
        print(f"Age   : {self.age}")
        print(f"Marks : {self.marks}")

student1 = Student("Ali", 20, 90)
student2 = Student("Sara", 21, 95)

student1.display()

print()

student2.display()
```
